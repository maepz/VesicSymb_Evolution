```python
import pandas as pd
import json
import numpy as np
from os import listdir

path="/home/maeperez/scratch/VesicSymb_Evolution/relax"
path_out="/home/maeperez/VesicSymb_Evolution/dowstream_analyses_and_data/"
```

#### results of RELAX using gene-specific trees ####

```python
#### get list of files for relax folder ####

genes=[]
for file in listdir(path):
    if file[-5:]=="relax":
        genes+=[str(file.split('.')[0])]
print(len(genes)) 
```

```python
#### open json file and get important results ####
pairwisecomps=['SymvsFL','RuvsFL','GivsFL','GivsRu','SymvsSUP05','SymvsHSym','SUP05vsHSym']
df=pd.DataFrame([],columns=['gene','ncodons']+[pfx+sfx for pfx in pairwisecomps for sfx in ['_k','_pval']])

failed=[]
for gene in sorted(genes):
    dic={}
    dic['gene']=gene
    print(gene)
    # get data from json files
    for comp in pairwisecomps:
#         print(comp)
        try :
            remote_file = open(path+'/'+gene+'_'+comp+'.RELAX.json')
            relax_json = json.load(remote_file)
            ncodons=int(relax_json['input']['number of sites'])
            pval=float(relax_json['test results']['p-value'])
            k=float(relax_json['test results']['relaxation or intensification parameter'])
            dic['ncodons']=ncodons
            dic[comp+'_pval']=pval
            dic[comp+'_k']=k
            remote_file.close()
        except:
            failed+=[[gene,comp]]
            continue
#     print(dic)
#     if 'GivsFL_k' not in dic.keys():
#         print '\t'+path+'/aligned_'+gene+'.fna'+'\\'
    df=df.append(dic, ignore_index=True)
df.drop(0,inplace=True)
```

```python
df.to_csv('/home/maeperez/VesicSymb_Evolution/dowstream_analyses_and_data/relax_results_outgroups.txt',header=True,index=False,sep='\t')
df
```
