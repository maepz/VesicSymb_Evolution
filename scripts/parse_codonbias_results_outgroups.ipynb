```python
from os import listdir
import pandas as pd
from Bio import SeqIO
import re
path='/home/maeperez/scratch/VesicSymb_Evolution/codon_bias/'
path_out="/home/maeperez/VesicSymb_Evolution/dowstream_analyses_and_data/"
```

```python

core=pd.read_csv('/home/maeperez/VesicSymb_Evolution/mauve_alignments/Outgroup/'+'nonrecomb_symb_and_outgroup_core_orthology.txt',
                header=0, sep='\t', index_col=None)
core
```

```python
core_dic={}
for col in core:
    core_dic.update(dict(list(zip(core[col].values,core['R.magnifica'].values))))
print(len(core_dic))
```

```python
cdc=pd.DataFrame([],columns=['GC', 'GC1', 'GC2', 'GC3', 'CDC', 'P(CDC)', 'coregene_id', 'ortho_id', 'locus_tag','ispseudo'])

## fill in outgroups genomes

for file in listdir(path):
    if file[-4:]=='.cat':
        gen=file.replace('.cat','')
        df=pd.read_csv(path+file,header=0,sep='\t')
#         ## Check cat file has same number of genes as input 
#         recs=[rec for rec in SeqIO.parse(path+gen+'.cds.fasta','fasta')]
#         if len(recs) != len(df):
#             print('something went wrong with CDC program')
#             print('#CDS in fasta:',len(recs),'\n#CDS if CDC',len(df))
#             continue
        if gen[-4:]=='.gbk':
            gen=gen[:-4]
        print(gen)
        df['coregene_id'], df['locus_tag'] = df['ID'].astype(str).str.split('|', 1).str
        df['coregene_id']=df['locus_tag'].map(core_dic)
        df['ncodons']=df['Length']/.3
        df['ortho_id']=df['locus_tag'].map(ortho_dic)
        df.drop(['ID'],axis=1,inplace=True)
        df['Genome']=gen.replace('.gbk','')
        cdc=cdc.append(df,ignore_index=True)
cdc_outgroup=cdc.sort_values('locus_tag')

cdc_outgroup.reset_index(inplace=True)
cdc_outgroup.drop(['index'],axis=1,inplace=True)
```

```python
## Add CDC table from other genomes

symb_cdc=pd.read_csv(path_out+'codonbias_results.csv',sep='\t',header=0,index_col=0)
symb_cdc
```

```python
cdc=symb_cdc.append(cdc_outgroup,ignore_index=True)
cdc
```

```python
cdc.to_csv(path_out+'codonbias_results_outgroups.csv',sep='\t')
print(set([val for val in cdc['Genome'].values]))
```
