```python
import os
import Bio
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.SeqFeature import SeqFeature, FeatureLocation
from Bio import AlignIO
from io import StringIO
import pandas as pd
import re
```

```python
mypath='../genomes/'
files=["A.diagonalis.gbk","A.gigas1.gbk","A.gigas2.gbk","A.mariana_ref.gbk","A.phaseoliformis.gbk","A.phaseoliformis_ref.gbk","A.southwardae.gbk","C.magnifica.gbk","C.pacifica.gbk","C.rectimargo.gbk","MK948426.gbk","MT528632.1.gbk","P.extenta.gbk","P.okutanii_ref.gbk","P.soyoae1.gbk","P.soyoae2.gbk","X.pliocardia.gbk","Bathy_mito.gbk","M.meretrix.gbk","M.lamarckii.gbk"]
onlyfiles = [mypath+f for f in files]
onlyfiles
```

```python
filedic={}
recs={}
path_dic={}
for file in onlyfiles:
    for record in SeqIO.parse(file,'genbank'):
        record.id='Contig_1'
        print(record.description)

        recs[record.name]={}
        path_dic[record.name]=file
      
        for feature in record.features:

            if feature.type!='CDS':
                continue

            recs[record.name].update({feature.qualifiers['gene'][0].upper():feature.qualifiers['locus_tag'][0]})

df=pd.DataFrame.from_dict(recs,orient='index')
df
```

```python
from Bio import Alphabet
from Bio.Alphabet import IUPAC,Gapped
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Align.Applications import MuscleCommandline
from Bio.codonalign import build
from Bio.Data.CodonTable import generic_by_id,unambiguous_dna_by_id
import subprocess
import sys

i=0
table5=unambiguous_dna_by_id[5]
table5.nucleotide_alphabet=IUPAC.ambiguous_dna
table5.protein_alphabet=IUPAC.extended_protein
table5.forward_table['AAN'] = "K"
table5.forward_table['GAY'] = "D"
table5.forward_table['YTA'] = "L"
table5.forward_table['ATY'] = "I"
table5.forward_table['TCY'] = "S"
table5.forward_table['GTY'] = "V"
table5.forward_table['GGY'] = "G"
table5.forward_table['CCY'] = "P"
table5.forward_table['GGR'] = "G"
table5.forward_table['ANN'] = "X"
table5.forward_table['NNN'] = "X"
table5.forward_table['NNA'] = "X"
table5.forward_table['NNG'] = "X"
table5.forward_table['GNN'] = "X"
table5.forward_table['NTG'] = "X"
table5.forward_table['GGN'] = "G"
table5.forward_table['TCN'] = "S"

for gene in df2.columns.values:
    i+=1
    j=0
    print(gene)
    fna=[]
    faa=[]
    for index in df2.index.values:
        j+=1
        locus_tag=df2[gene][index]
        record = SeqIO.read(path_dic[index], "genbank")
        for feature in record.features:
            if feature.type != 'CDS':
                continue
            if feature.qualifiers['locus_tag'][0]==locus_tag:
                
                ### create SeqRecord object ####
                DNAseq=feature.location.extract(record).seq
                AAseq=Seq(feature.qualifiers['translation'][0],alphabet=Gapped(IUPAC.extended_protein))
                header='|'.join([index,locus_tag])

                fna_seq = SeqRecord(DNAseq,id=header)
                faa_seq = SeqRecord(AAseq,id=header)

                fna.append(fna_seq)
                faa.append(faa_seq)
#                 if j==1:
#                     fna.append(Svel[gene]['fna'])
#                     faa.append(Svel[gene]['faa'])

    faa=[f for f in sorted(faa, key=lambda x : x.id)]
    fna=[f for f in sorted(fna, key=lambda x : x.id)]
    SeqIO.write(faa, '../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/'+gene+'.faa', "fasta")

    muscle_cline = MuscleCommandline('/cvmfs/soft.computecanada.ca/easybuild/software/2017/avx512/Compiler/gcc7.3/muscle/3.8.31/bin/muscle',input='../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/'+gene+'.faa', out='../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/aligned_'+gene+'.faa')
    muscle_cline()
    align = AlignIO.read('../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/aligned_'+gene+'.faa', "fasta",alphabet=Gapped(IUPAC.extended_protein))

#     stdout, stderr = muscle_cline()
#     align = AlignIO.read(StringIO(stdout), "fasta")
    print(align)

    
#     #### run MUSCLE ####
#     muscle_cline = MuscleCommandline(clwstrict=True)
#     child = subprocess.Popen(str(muscle_cline),
#                      stdin=subprocess.PIPE,
#                      stdout=subprocess.PIPE,
#                      stderr=subprocess.PIPE,
#                      universal_newlines=True,
#                      shell=(sys.platform!="win32"))
#     SeqIO.write(faa, child.stdin, "fasta")
#     child.stdin.close()
#     align = AlignIO.read(child.stdout, "clustal",alphabet=Gapped(IUPAC.extended_protein))
    aln = MultipleSeqAlignment([f for f in sorted(align, key=lambda x : x.id)])
    print(aln)
    print(fna)
    codon_aln = build(aln, fna,complete_protein=False,max_score=30,codon_table=table5)

    try:
        codon_aln = build(aln, fna,complete_protein=False,max_score=30,codon_table=table5)
        print(codon_aln)
        print(len(codon_aln))
    except:
        continue
    

    SeqIO.write(codon_aln, '../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/'+gene+'.fna', "fasta")

    n=0
    names={}
    for seq in codon_aln:
        seq.id=seq.id.split('|')[0]
        seq.name=seq.id
        names[n]=seq.name.replace('.','')
        n+=1
    if i==1:
        cat=codon_aln
    if i>1:
        cat.sort()
        codon_aln.sort()
        cat+=codon_aln
for i in range(len(cat)):
    cat[i].id=names[i]
print(cat)

SeqIO.write(cat, '../mauve_alignments/Mitochondria/MITO_Mlamarckii/Codon_aln/cat_mitoch_genes.fna', "fasta")
```

```python
### This script if the gbk file needs locus_tags added ###

filedic={}
recs={}
print(onlyfiles[-1])
file=onlyfiles[-1]
for record in SeqIO.parse(file,'genbank'):
    record.id='Contig_1'
    print(record.description)


    locus_tag_prefix='Mlam'
    print(locus_tag_prefix)  
   ### create source feature 
    source_feature=SeqFeature(FeatureLocation(1, len(record), strand=1), type="source")
    source_feature.qualifiers['organism']=[' '.join(record.description.split(' ')[:2])]
    source_feature.qualifiers['organelle']= ['mitochondrion']
    source_feature.qualifiers['mol_type']= ['genomic DNA']

    i=0
    recs[record.name]={}
    new_features=[]
    new_features+=[source_feature]         

    for feature in record.features:

        if feature.type=='source':
            continue
        # edit feature qualifiers for non genes
        if feature.type=='gene':
            continue
        else:
            i+=1
#                 print(feature.qualifiers['locus_tag'])
            feature.qualifiers['locus_tag']=[locus_tag_prefix+'_'+"{:02d}".format(i)]
            feature.qualifiers.pop('standard_name', None)
            if feature.type=='tRNA':
#                     print(feature)
                feature.qualifiers['gene']=[feature.qualifiers['product'][0]] 
            if feature.type=='rRNA':
#                     print(feature)
                feature.qualifiers['gene']=[feature.qualifiers['product'][0]] 

            if feature.type=='CDS':
#                     print(feature.qualifiers.keys())
#                     print(feature.qualifiers['locus_tag'])
                feature.qualifiers['transl_table']=[5]
                feature.qualifiers['gene'][0]=feature.qualifiers['gene'][0].upper()
#                     feature.qualifiers['protein_id']=['gnl|CRYetCie|'+feature.qualifiers['locus_tag'][0]]
                recs[record.name].update({feature.qualifiers['gene'][0]:feature.qualifiers['locus_tag'][0]})


                if feature.qualifiers['gene'][0]=='ND3':
                    feature.qualifiers['product']=["NADH dehydrogenase subunit 3"]
                if feature.qualifiers['gene'][0]=='ND4':
                    feature.qualifiers['product']=["NADH dehydrogenase subunit 4"]

                try:
                    feature.qualifiers['translation']=[str(feature.extract(record).translate(table=5).seq)]
                except:
                    print('!!! issue with cds')
                    pass


            if 'note' in feature.qualifiers.keys():
                if 'Derived using Geneious Prime' in feature.qualifiers['note'][0]:
                    feature.qualifiers.pop('note', None)

            ### create gene feature 
            gene_feature=SeqFeature(feature.location, type="gene")
            gene_feature.type='gene'
            gene_feature.qualifiers={'locus_tag':feature.qualifiers['locus_tag'],'gene':feature.qualifiers['gene']}
            new_features+=[gene_feature,feature]
            ### create tbl
            #Gene
            start=int(feature.location.start)+1
            end=int(feature.location.end)
            if feature.strand<0:
                start=int(feature.location.end)
                end=int(feature.location.start)+1


    ### write output file gbk and fasta
    record.features=new_features
    SeqIO.write(record,file[:-3]+'.gbk','genbank')
#         SeqIO.write(record,mypath2+file[:-3]+'.fasta','fasta')


df2=df.append(pd.DataFrame.from_dict(recs,orient='index'))
df2
```
