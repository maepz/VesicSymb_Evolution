```python
def add_labels_to_tree(input_tree_file):
    dic={'MS2017':'{fl}','SP60':'{fl}','Rkil':'{gigas}','COSY':'{gigas}','Vdia':'{gigas}','Vext':'{gigas}','Vgig':'{gigas}','Rmag':'{ruthia}','Rple':'{ruthia}','Rsou':'{ruthia}','Rpac':'{ruthia}','Rrec':'{ruthia}','Rpha':'{ruthia}'}
    dic2={'MS2017':'{fl}','SP60':'{fl}','Rkil':'{sym}','COSY':'{sym}','Vdia':'{sym}','Vext':'{sym}','Vgig':'{sym}','Rmag':'{sym}','Rple':'{sym}','Rsou':'{sym}','Rpac':'{sym}','Rrec':'{sym}','Rpha':'{sym}'}
    tree=Tree(input_tree_file,format=1)
    tree2=Tree(input_tree_file,format=1)

    for leaf in tree.iter_leaf_names():
        for item in leaf.split('_')[1:]:
            try:
                #tree1
                new_leaf_name=leaf+dic[item]
                node=tree&leaf
                node.name=new_leaf_name

                #tree2
                new_leaf_name2=leaf+dic2[item]
                node2=tree2&leaf
                node2.name=new_leaf_name2

            except KeyError:
                continue


    for node in tree.traverse():
        if len(set([l.name.split('_')[-1] for l in node.get_leaves()]))==1:
            if len(node.name) ==0:
                node.name=[l.name.split('_')[-1] for l in node.get_leaves()][0]

#     print tree.get_ascii(show_internal=True)
    tree1_output=tree.write(format=1)


    for node in tree2.traverse():
        if len(set([l.name.split('_')[-1] for l in node.get_leaves()]))==1:
            if len(node.name) ==0:
                node.name=[l.name.split('_')[-1] for l in node.get_leaves()][0]

#     print tree2.get_ascii(show_internal=True)
    tree2_output= tree2.write(format=1)

    return tree.get_ascii(show_internal=True),tree1_output,tree2.get_ascii(show_internal=True), tree2_output
```

```python
def add_labels_to_treev2(input_tree_file):
    dic={'MS2017':'{fl}','SP60':'{fl}','Rkil':'{gigas}','COSY':'{gigas}','Vdia':'{gigas}','Vext':'{gigas}','Vgig':'{gigas}','Rmag':'{ruthia}','Rple':'{ruthia}','Rsou':'{ruthia}','Rpac':'{ruthia}','Rrec':'{ruthia}','Rpha':'{ruthia}'}
    dic2={'MS2017':'{fl}','SP60':'{fl}','Rkil':'{sym}','COSY':'{sym}','Vdia':'{sym}','Vext':'{sym}','Vgig':'{sym}','Rmag':'{sym}','Rple':'{sym}','Rsou':'{sym}','Rpac':'{sym}','Rrec':'{sym}','Rpha':'{sym}'}
    tree=Tree(input_tree_file,format=1)
    tree2=Tree(input_tree_file,format=1)

    for leaf in tree.iter_leaf_names():
        for item in leaf.split('_')[1:]:
            try:
                #tree1
                new_leaf_name=leaf+dic[item]
                node=tree&leaf
                node.name=new_leaf_name

                #tree2
                new_leaf_name2=leaf+dic2[item]
                node2=tree2&leaf
                node2.name=new_leaf_name2

            except KeyError:
                continue
    treeroot=[]
    for leaf in tree.get_leaf_names():
        if leaf.split('_')[0] in ['Bathy','SUP05']:
            treeroot+=[leaf]
#     print list(treeroot)
#     print tree
    ancestor = tree.get_common_ancestor(list(treeroot))
    try:
        tree.set_outgroup(ancestor)
    except:
        tree.set_outgroup(tree&treeroot[0])
    for node in tree.traverse():
        if len(set([l.name.split('_')[-1] for l in node.get_leaves()]))==1:
            if len(node.name) ==0:
                node.name=[l.name.split('_')[-1] for l in node.get_leaves()][0]

#     print tree.get_ascii(show_internal=True)
    tree1_output=tree.write(format=1)

    ancestor = tree2.get_common_ancestor(list(treeroot))
    try:
        tree2.set_outgroup(ancestor)
    except:
        tree2.set_outgroup(tree2&treeroot[0])
    for node in tree2.traverse():
        if len(set([l.name.split('_')[-1] for l in node.get_leaves()]))==1:
            if len(node.name) ==0:
                node.name=[l.name.split('_')[-1] for l in node.get_leaves()][0]

#     print tree2.get_ascii(show_internal=True)
    tree2_output= tree2.write(format=1)

    return tree.get_ascii(show_internal=True),tree1_output, tree2.get_ascii(show_internal=True), tree2_output
```
