```python
import pandas as pd
######## read and parse ortholog file #########
genomes=["SUP05","R.magnifica","Bathy_complete_with_CRISPRannot","Bathysepcomplete_AP013042","Bathyther_MIQH01","Bathy2_CAEB01","Bathyazo_CAESAP02","Sponge_JYIN01","ThioNP1complete_CP023860","ThiosingularisPS1complete_CP006911","Thioautotrophicus_DQLA01","Thioperditus_PNQY01","Thiopontius_JACNGB01","ThioTMED218_NHJP01"]
locus_tag_prefixes=[]
path_to_files='../mauve_alignments/Outgroup/'

#pfx=dict(zip(locus_tag_prefixes,genomes))
idx_dict=dict(zip(range(len(genomes)),genomes))

OUTGROUP_raw=pd.read_csv(path_to_files+'OUTGROUP_id35cov51.orthologs',sep='\t',header=None,names=genomes)
OUTGROUP=pd.DataFrame([],columns=genomes, index=OUTGROUP_raw.index)

for idx in OUTGROUP_raw.index:
    for col in list(OUTGROUP_raw):
        if type(OUTGROUP_raw.loc[idx][col])!=float:
            column=OUTGROUP_raw.loc[idx][col].split(':')[0]
            value=OUTGROUP_raw.loc[idx][col].split(':')[1]
            OUTGROUP.loc[idx][idx_dict[int(column)]]=value
```

```python
            
good_genomes=["SUP05","R.magnifica","Bathy_complete_with_CRISPRannot","Bathysepcomplete_AP013042","Bathyther_MIQH01","Bathyazo_CAESAP02","Sponge_JYIN01","ThioNP1complete_CP023860","ThiosingularisPS1complete_CP006911","Thioautotrophicus_DQLA01","Thioperditus_PNQY01","Thiopontius_JACNGB01"] 
OUTGROUP=OUTGROUP[good_genomes]
OUTGROUP['Count']=OUTGROUP[good_genomes].count(axis=1)
# ALL.to_csv('Orthology_id35cov51.txt',header=True,sep='\t')  

CORE_OUTGROUP=OUTGROUP[OUTGROUP['Count']==len(OUTGROUP.columns)-1]
           
OUTGROUP.to_csv(path_to_files+'OUTGROUP_Orthology_id35cov51.txt',header=True,sep='\t')  
CORE_OUTGROUP.to_csv(path_to_files+'OUTGROUP_CoreOrthology_id35cov51.txt',header=True,sep='\t') 
```

```python
########### CORE: retreive sequences from GBK, align aa, backtranslate and output fna; 715 sequences ##########
from Bio import AlignIO
from Bio.codonalign import build
from Bio.Alphabet import Gapped,IUPAC
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Align import MultipleSeqAlignment
from Bio.Alphabet import generic_dna, generic_protein
from Bio.Align.Applications import MuscleCommandline,ClustalwCommandline
import subprocess
import sys
import os

# os.mkdir('../mauve_alignments/Outgroup/faa')
# os.mkdir('../mauve_alignments/Outgroup/fna')


goodfiles={'Bathy':'../genomes/Bathy_complete_with_CRISPRannot.gbk',\
'R.fausta':'../genomes/R.fausta.gbk',\
'R.magnifica':'../genomes/R.magnifica.gbk',\
'R.pacifica':'../genomes/R.pacifica_one_contig_circular_fully_annotated.gbk',\
'R.phaseoliformis':'../genomes/R.phaseoliformis_8_contigs_fully_annotated.gbk',\
'R.pliocardia':'../genomes/R.pliocardia_one_contig_circular_fully_annotated.gbk',\
'R.rectimargo':'../genomes/R.rectimargo_one_contig_circular_fully_annotated.gbk',\
'R.southwardae':'../genomes/R.southwardae_39_contigs_fully_annotated.gbk',\
'SUP05':'../genomes/SUP05.gbk',\
'V.diagonalis':'../genomes/V.diagonalis_one_contig_circular_fully_annotated.gbk',\
'V.extenta':'../genomes/V.extenta_one_contig_circular_fully_annotated.gbk',\
'V.gigas1':'../genomes/V.gigas1_one_contig_circular_fully_annotated.gbk',\
'V.gigas2':'../genomes/V.gigas2_one_contig_circular_fully_annotated.gbk',\
'V.marissinica':'../genomes/V.marissinica_withlocustags.gbk',\
'V.okutanii':'../genomes/V.okutanii_with_locus_tags.gbk',\
'V.soyoae1':'../genomes/V.soyoae1_one_contig_circular_fully_annotated.gbk',\
'V.soyoae2':'../genomes/V.soyoae2_one_contig_circular_fully_annotated.gbk',\
'Bathysepcomplete_AP013042':'../genomes/additional_genomes/Bathysepcomplete_AP013042.gbk', \
'Bathyther_MIQH01':'../genomes/additional_genomes/Bathyther_MIQH01.gbk', \
'Bathyazo_CAESAP02':'../genomes/additional_genomes/Bathyazo_CAESAP02.gbk', \
'Sponge_JYIN01':'../genomes/additional_genomes/Sponge_JYIN01.gbk', \
'ThioNP1complete_CP023860':'../genomes/additional_genomes/ThioNP1complete_CP023860.gbk', \
'ThiosingularisPS1complete_CP006911':'../genomes/additional_genomes/ThiosingularisPS1complete_CP006911.gbk', \
'Thioautotrophicus_DQLA01':'../genomes/additional_genomes/Thioautotrophicus_DQLA01.gbk', \
'Thioperditus_PNQY01':'../genomes/additional_genomes/Thioperditus_PNQY01.gbk', \
'Thiopontius_JACNGB01':'../genomes/additional_genomes/Thiopontius_JACNGB01.gbk', \
'ThioTMED218_NHJP01':'../genomes/additional_genomes/ThioTMED218_NHJP01.gbk', \
'Bathy2_CAEB01':'../genomes/additional_genomes/Bathy2_CAEB01.gbk'}

CORE=pd.read_csv(path_to_files+'nonrecomb_symb_and_outgroup_core_orthology.txt',header=0,sep='\t')
CORE
```

```python
from Bio import SeqIO
from Bio.Data.CodonTable import generic_by_id,unambiguous_dna_by_id


table11=unambiguous_dna_by_id[11]
table11.nucleotide_alphabet=IUPAC.ambiguous_dna
table11.protein_alphabet=IUPAC.extended_protein
table11.forward_table['ATN'] = "I"
table11.forward_table['ACN'] = "T"
table11.forward_table['GTN'] = "V"
table11.forward_table['GCN'] = "A"
table11.forward_table['TCN'] = "S"
table11.forward_table['CTN'] = "L"
table11.forward_table['CCN'] = "P"
table11.forward_table['CGN'] = "R"
table11.forward_table['GGN'] = "G"
table11.forward_table['NCA'] = "X"

for i in range(len(CORE)):

    gene=CORE.iloc[i]
    gene_name=CORE.iloc[i]['R.magnifica']
    if gene_name.split('_')[1][0]=='R':
        continue
    fna=[]
    faa=[]
    print(i,gene_name)
    for sample in sorted(gene.index):
        for record in SeqIO.parse(goodfiles[sample],'genbank'):
            for feature in record.features:
                if feature.type=='CDS':
                    locus_tag = feature.qualifiers['locus_tag'][0]
                    if locus_tag == CORE.iloc[i][sample]:
                        
                        ### create SeqRecord object ####
                        DNAseq=feature.location.extract(record).seq
                        if locus_tag == gene_name:
                            print(len(DNAseq))
                        try:
                            AAseq=Seq(feature.qualifiers['translation'][0],alphabet=Gapped(IUPAC.extended_protein))
                        except KeyError:
                            print(feature)
                            AAseq=feature.extract(record).translate(11)
                            AAseq.id=locus_tag
#                             print(AAseq.seq)
                        header='|'.join([sample,CORE.iloc[i][sample]])

                        fna_seq = SeqRecord(DNAseq,id=header+' /transl_table=11')
                        faa_seq = SeqRecord(AAseq,id=header+' /transl_table=11')
                        
                        fna.append(fna_seq)
                        faa.append(faa_seq)                  

    faa=[f for f in sorted(faa, key=lambda x : x.id)]
    fna=[f for f in sorted(fna, key=lambda x : x.id)]

    SeqIO.write(faa, path_to_files+'faa/'+gene_name+'.faa', "fasta")
    
    if len(fna[0])>2000:
        SeqIO.write(fna, path_to_files+'fna/'+gene_name+'.fna', "fasta")
        error+=[i]
        continue

    
#     muscle_cline = MuscleCommandline('/cvmfs/soft.computecanada.ca/easybuild/software/2017/avx512/Compiler/gcc7.3/muscle/3.8.31/bin/muscle',input='fna/'+gene_name+'.fasta', out='fna/aligned_'+gene_name+'.faa')
#     muscle_cline()
#     align = AlignIO.read('fna/aligned_'+gene_name+'.faa', "fasta",alphabet=Gapped(IUPAC.extended_protein))

#     stdout, stderr = muscle_cline()
#     align = AlignIO.read(StringIO(stdout), "fasta")
#     print(align)

    ### run MUSCLE ####
    muscle_cline = MuscleCommandline(clwstrict=True)
    child = subprocess.Popen(str(muscle_cline),
                             stdin=subprocess.PIPE,
                             stdout=subprocess.PIPE,
                             stderr=subprocess.PIPE,
                             universal_newlines=True,
                             shell=(sys.platform!="win32"))
    SeqIO.write(faa, child.stdin, "fasta")
    child.stdin.close()
    align = AlignIO.read(child.stdout, "clustal",alphabet=Gapped(IUPAC.extended_protein))
    aln = MultipleSeqAlignment([f for f in sorted(align, key=lambda x : x.id)])
    try:
        codon_aln = build(aln, fna,complete_protein=False,max_score=1000,codon_table=table11)
        SeqIO.write(codon_aln, path_to_files+'fna/aligned_'+gene_name+'.fna', "fasta")
    except Error:
        SeqIO.write(fna, path_to_files+'fna/'+gene_name+'.fna', "fasta")
        error+=[i]
        continue
#     SeqIO.write(aln, path_to_files+'faa/aligned_'+gene_name+'.faa', "fasta")
```
